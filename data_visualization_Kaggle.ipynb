```python
# https://www.kaggle.com/alexisbcook/choosing-plot-types-and-custom-styles

# "the above link shows the type of graphs"
# Starting witht the name of Almighty Allah
# this is the kaggle course for data visualization
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("done")
```

```python
# Path of the file to read
df= pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv", delimiter = ",")
df= df.drop(columns=['country','continent'])
df.head()
```

```python
# setting width and height of the figure
plt.figure(figsize= (15,5))

# Line chart
sns.lineplot(data= df)


```

***Bar chars & heat maps***

```python
# Bar chart
df= df.head(100)
plt.figure(figsize= (10,6))

# Add title
plt.title("Example of a bar chart")

sns.barplot(x=df.index, y= df['beer_servings'])

# Add label for vertical axis

plt.ylabel("y-axis tile")
```

```python
#Heatmap
# setting figure size
plt.figure(figsize=(14,7))

# Add title

plt.title("Example of Heatmap")

# Heat map

sns.heatmap(data = df, annot = True)

# Add label for horizontal azix

plt.xlabel("x-axis")
```

***Scatter plots***

```python
sns.scatterplot(x= df['beer_servings'], y = df['wine_servings'], hue= df['spirit_servings'])
```

```python
sns.lmplot(x = 'beer_servings' , y = 'wine_servings', hue = 'spirit_servings', data=df)
```

```python
sns.swarmplot(x= df['beer_servings'], y= df['wine_servings'])
```

***Distributions***

```python
#Histograms

sns.distplot( a= df['wine_servings'], kde = False)
```

```python
# KDE plot

sns.kdeplot(data= df['wine_servings'], shade= True)
```

```python
# 2D KDE plot
sns.jointplot(x= df['wine_servings'], y= df['beer_servings'], kind= "kde")
```
