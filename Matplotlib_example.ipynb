```python
#Plotting graphs using python libraries
import numpy as no
import pandas as pd
import matplotlib.pyplot as plt

```

```python
plt.plot([1,3,2,4])
plt.show()
```

```python
# Points have x and y values; add title and axis labels
plt.plot([1,2,3,4],[1,4,9,16])
plt.title('Test Plot', fontsize = 14, color='g')
plt.xlabel('values of x')
plt.ylabel('values of y')
plt.show()
```

```python
#Change figure size. plot red dots; set axis scales x: 0-6 and y: 0-20
plt.figure(figsize=(3,5))  # 3 inches wide x 5 inches tall
plt.plot([1,2,3,4], [1,4,9,16], 'ro') # red-o
plt.axis([0,6,0,20]) # [xmin, xmax, ymin, ymax]
plt.annotate('square it', (3,6))  # annotate meaning : add word or test
plt.show()
```

```python
# Two sets of 10 random plotted
d = { 'Red O': np.random.rand(10),
      'Grn X': np.random.rand(10)}
df= pd.DataFrame(d)
df.plot(style=['ro','gx'])
plt.show()
```

```python
# Time series - six months of random floats
ts = pd.Series(np.random.randn(180), index=pd.date_range('1/1/2018', periods= 180))
df = pd.DataFrame(np.random.randn(180, 3), index=ts.index, columns=list('ABC'))
df.cumsum().plot()
plt.show()
```

```python
#Random dots in a scatter
N =50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30* np.random.rand(N))**2   # 0 to 15 point radii
plt.scatter(x,y, s= sizes, c= colors, alpha = 10) # alpha lower value will dim the image
plt.show()
```

```python
#Load csv file and show multiple chart types

df = pd.read_csv('Fremont_weather.txt')
print(df)
plt.bar(df['month'], df['record_high'], color='r')
plt.bar(df['month'], df['record_low'], color='c')
plt.plot(df['month'], df['avg_high'], color='k')
plt.plot(df['month'], df['avg_low'], color='b')
plt.legend()
plt.show()
```

```python
'''
Subplots, part 1
221 = top left subplot
222 = top right subplot
223 = bottom left subplot
224 = bottom right subplot
'''
fig = plt.figure()
fig.suptitle('My SubPlots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224) 
plt.show()
```

```python
#Subplots, part 2
fig, plots = plt.subplots(2, sharex=True)
fig.suptitle('Sharing X axis')
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x, y, color='r')
plots[1].scatter(x, y)
plt.show()
```
